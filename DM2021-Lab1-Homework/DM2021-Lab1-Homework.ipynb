```python
import re
from bs4 import BeautifulSoup
import string
import nltk
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
from nltk.stem import WordNetLemmatizer
from nltk import word_tokenize, pos_tag

# nltk.download('stopwords')
stop = stopwords.words('english')
```

##  preprocessing
* string to lower
* remove punctuation
* use PorterStemmer
* remove stop words

```python
def preprocessor(text):
    # remove HTML tags
    text = text.lower()
    text = text.translate(str.maketrans('', '', string.punctuation))
    text = text.strip()
    return text

def tokenizer_stem_nostop(text):
    text = preprocessor(text)
    porter = PorterStemmer()
    return [porter.stem(w) for w in re.split('\s+', text.strip()) \
            if w not in stop and re.match('[a-zA-Z]+', w)]
```

```python
X['unigrams'] = X['text'].apply(lambda x: dmh.tokenize_text(x))
X['tokens'] = X['text'].apply(lambda x: tokenizer_stem_nostop(x))
```

```python
X['tokens']
```

```python
X['tokens'] = X['tokens'].apply(lambda x: ' '.join(x))
```

## tfidf and visualization

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfTransformer, TfidfVectorizer
tfidf_vectorize = TfidfVectorizer()

# here 
X_tfidf = tfidf_vectorize.fit_transform(X.tokens)
```

```python
tfidf_vectorize.get_feature_names()[0:10]
```

```python
df_todraw = pd.DataFrame(X_tfidf.toarray(), columns = tfidf_vectorize.get_feature_names())

threshold = 0.8
keep_columns = df_todraw.max()>=threshold
df_todraw = df_todraw[df_todraw.columns[keep_columns.values==True]]
df_todraw = df_todraw[:20]
for term in df_todraw.columns:
    df_todraw.rename({term : f'term_{term[:10]}'}, axis=1, inplace=True)
for i in range(20):
    df_todraw.rename({i : f'doc_{i}'}, axis=0, inplace=True)
    
print(df_todraw.shape)
df_todraw
plt.subplots(figsize=(9, 7))
ax = sns.heatmap(df_todraw,
                 cmap="PuRd",
                 vmin=0, vmax=1, annot=True)
```

## Dimensionality Reduction

```python
from sklearn.decomposition import PCA
X_reduced = PCA(n_components = 2).fit_transform(X_tfidf.toarray())
col = ['red', 'blue']

# plot
fig = plt.figure(figsize = (25,10))
ax = fig.subplots()
categories = [0,1]

for c, category in zip(col, categories):
    xs = X_reduced[X['score'] == category].T[0]
    ys = X_reduced[X['score'] == category].T[1]
   
    ax.scatter(xs, ys, c = c, marker='o')

ax.grid(color='gray', linestyle=':', linewidth=2, alpha=0.2)
ax.set_xlabel('\nX Label')
ax.set_ylabel('\nY Label')

plt.show()
```
